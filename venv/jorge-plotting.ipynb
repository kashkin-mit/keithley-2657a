```python
import json
import os
import matplotlib.pyplot as plt
```

```python
data_dir = 'C://Users//kashkin//Desktop//keithley data//'
```

```python
files = ['current up to 0.12 A for sample .6 3_ 2022_11_23-18_43_50', 'current up to 0.12 A for sample .6 2_ 2022_11_23-18_42_26','current up to 0.12 A for sample .6_ 2022_11_23-18_41_37']
```

```python
for file in files:
    with open(data_dir + file + '.json', 'r') as f:
        data = json.load(f)

    plt.plot(data['input_currents'],data['measured_voltages'],label=file[25:36])

plt.legend()
plt.xlabel('Current (A)')
plt.ylabel('Voltage (V)')
plt.show()
```
